# Análise das Viagens a Serviço no Governo Federal no ano de 2024

Neste projeto, foram utilizados dados de viagens a serviço disponíveis no Portal da Transparência do Governo Federal (https://portaldatransparencia.gov.br/). Através deste portal, é possível acessar informações detalhadas sobre diárias e passagens de viagens governamentais realizadas por servidores públicos do Poder Executivo Federal, incluindo dados sobre destinos, custos, tipo de viagem, órgãos envolvidos e muito mais.

O objetivo deste projeto é analisar os gastos do governo federal com viagens a serviço através da análise dos custos com diárias, passagens e outros gastos, além de outras informações relacionadas a viagens governamentais, como cargos envolvidos, destinos, número de viagens, duração média das viagens. Para realizar a análise desses dados, foram empregadas as tecnologias Python, com a biblioteca Pandas para manipulação e análise dos dados, e Matplotlib para a visualização gráfica das informações.

```python
# Caminho de entradas e saídas dos dados
caminho_dados_viagens = './2024_Viagem.csv'
caminho_dados_passagens = './2024_Passagem.csv'
caminho_saida_tabela = './output/tabela_2024.xlsx'
caminho_saida_barra = './output/grafico_barra.png'
caminho_saida_dispersao = './output/grafico_dispersao.png'
```

```python
import pandas as pd

# Options para mostrar todas as colunas e apenas duas casas decimais em tipos float.
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)

# Lendo o arquivo csv.
df_viagens = pd.read_csv(caminho_dados_viagens, encoding='Windows-1252', sep=';', decimal=',')

# Criando coluna 'Despesas' com a soma de valores de diárias, passagens e outros gastos.
df_viagens['Despesas'] = df_viagens['Valor diárias'] + df_viagens['Valor passagens'] + df_viagens['Valor outros gastos']

# Reescrevendo a coluna 'Cargo' de forma que os valores nulos sejam substituídos por 'NÃO IDENTIFICADO'.
df_viagens['Cargo'] = df_viagens['Cargo'].fillna('NÃO IDENTIFICADO')

# Transformando os valores das colunas que representam datas para o formato datetime.
df_viagens['Período - Data de início'] = pd.to_datetime(df_viagens['Período - Data de início'], format='%d/%m/%Y')
df_viagens['Período - Data de fim'] = pd.to_datetime(df_viagens['Período - Data de fim'], format='%d/%m/%Y')

# Adicionando a coluna mês da viagem pegando somente o mês da coluna 'Período - Data de início'.
df_viagens['Mês da viagem'] = df_viagens['Período - Data de início'].dt.month_name()

# Adicionando a coluna dias de viagem fazendo a subtração da data de fim da viagem com a data de início.
df_viagens['Dias de viagem'] = (df_viagens['Período - Data de fim'] - df_viagens['Período - Data de início']).dt.days
```

```python
# Criando tabela consolidada
df_viagens_consolidado = (
    df_viagens.groupby('Cargo').agg(
        despesas_media=('Despesas', 'mean'),
        duracao_media=('Dias de viagem', 'mean'),
        despesas_totais=('Despesas', 'sum'),
        destino_mais_frequente=('Destinos', pd.Series.mode),
        n_viagens=('Nome', 'count')
    ).reset_index().sort_values(by='despesas_totais', ascending=False)
)

# Filtrando tabela consolidada por cargos relevantes (> 1% das viagens)
df_cargos = df_viagens['Cargo'].value_counts(normalize=True).reset_index()
df_cargos.columns = ['Cargo', 'proportion']
cargos_relevantes = df_cargos.loc[df_cargos['proportion'] > 0.01, 'Cargo']
filtro = df_viagens_consolidado['Cargo'].isin(cargos_relevantes)

# Tabela final
df_final = df_viagens_consolidado[filtro].sort_values(by='n_viagens', ascending=False)

# Salvando a tabela final
df_final.to_excel(caminho_saida_tabela)
```

```python
import matplotlib.pyplot as plt

# Definindo estilo global para gráficos
plt.rcParams['figure.figsize'] = (16, 6)
plt.rcParams['axes.facecolor'] = '#180840'
plt.rcParams['figure.facecolor'] = 'whitesmoke'
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.color'] = 'gray'
plt.rcParams['grid.linestyle'] = '--'
plt.rcParams['grid.linewidth'] = 0.5

# Criando gráfico com determinada dimensão
fig, ax = plt.subplots()

# Definindo gráfico do tipo barra e quais colunas serão usadas
ax.barh(df_final['Cargo'], df_final['n_viagens'], color='#5437a3')

# Invertendo ordem das barras
ax.invert_yaxis()

# Colocando título superior
fig.suptitle('Viagens por cargo público (2024)')

# Colocando label para o eixo x
plt.xlabel('Número de viagens')

# Salvando gráfico
plt.savefig(caminho_saida_barra, bbox_inches='tight')
```

Em seguida, decidi exibir um gráfico de dispersão para mostrar a relação entre as duas variáveis quantitativas "Dias de Viagem" e "Despesas", com finalidade de visualizar como as variáveis se correlacionam, identificar padrões e dados que se destacam ou estão distantes da maioria dos outros pontos.

```python
# Criando gráfico de dispersão com dados de dias de viagem e despesas
fig, ax = plt.subplots()

# Definindo gráfico de dispersão e quais colunas serão usadas
ax.scatter(df_viagens['Dias de viagem'], df_viagens['Despesas'], c='#5437a3')

# Colocando título superior
ax.set_title('Dias de Viagem por Despesas')

# Colocando label para eixo x e y
ax.set_xlabel('Dias de Viagem')
ax.set_ylabel('Despesas')

# Salvando gráfico
plt.savefig(caminho_saida_dispersao, bbox_inches='tight')
```

A análise de dispersão sugere que, embora exista uma relação entre o número de dias de viagem e as despesas, a distribuição das viagens apresenta variações significativas, com algumas se destacando por custos elevados, mesmo em viagens de curta duração. Viagens com despesas extremamente altas (próximas de R$ 300.000) foram identificadas como outliers no gráfico de dispersão. Essas viagens merecem uma investigação mais detalhada para entender as razões de seus custos elevados, especialmente considerando que os destinos eram relativamente próximos.
